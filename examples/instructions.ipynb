The function below takes base64encoded Word document and converts it as a PDF

```python
from pypaperflow.word2html import convert_word_to_pdf
import base64

base64_encoded_word = base64.b64encode(open("./files/resume.docx", "rb").read()).decode("utf-8")
pdf_base64 = await convert_word_to_pdf(word_file=base64_encoded_word, output_path="./files/word_output.pdf")
```

Now we take an HTLM file and convert it to a PDF

```python
from pypaperflow.html2pdf import create_pdf_from_html

html_content = open("/Users/bselleslagh/Downloads/test.html", "r").read()
# Directly save the PDF to a file
pdf_base64 = await create_pdf_from_html(html_content, output_path="./files/html_output.pdf")

# Return the PDF as a base64 encoded string, then decode it and save it to a file
html_content = open("/Users/bselleslagh/Downloads/test.html", "r").read()
pdf_base64 = await create_pdf_from_html(html_content)

with open("./files/html_output_test.pdf", "wb") as f:
    f.write(base64.b64decode(pdf_base64))
```
